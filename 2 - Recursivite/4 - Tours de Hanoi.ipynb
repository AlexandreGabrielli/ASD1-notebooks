```python
def Hanoi(Towers,Orig,Dest,Via,n):
    if n > 0:
        Hanoi(Towers,Orig,Via,Dest,n-1)
        Dest.append(Orig.pop())
        affichageHanoi(Towers)
        Hanoi(Towers,Via,Dest,Orig,n-1)
```
